# Desafio DIO de redução de dimensionalidade

Este é um desafio do Bootcamp da DIO

```python
from PIL import Image


# converte uma imagem colorida para níveis de cinza entre 0 e 255
def colorida_para_cinza(caminho_imagem):
    
    im = Image.open(caminho_imagem)
    im = im.convert('RGB')    # garantia de que está no formato RGB
    largura, altura = im.size
    im_cinza = Image.new('L', (largura, altura))   # cria uma imagem vazia em tons de cinza
    
    for x in range(largura):
        for y in range(altura):
            r, g, b = im.getpixel((x, y))
            
            cinza = int(0.2989 * r + 0.5870 * g + 0.1140 * b)
            im_cinza.putpixel((x, y), cinza)
    
    return im_cinza


# converte uma imagem em tons de cinza para uma imagem binarizada 0 ou 255
def cinza_para_binario(imagem_cinza, limiar=128):
    
    largura, altura = imagem_cinza.size
    imagem_binaria = Image.new('1', (largura, altura))   # cria uma imagem binarizada 1 bit
    
    for x in range(largura):
        for y in range(altura):
            
            cinza = imagem_cinza.getpixel((x, y))
            
            valor_binario = 255 if cinza > limiar else 0
            imagem_binaria.putpixel((x, y), valor_binario)
        
    return imagem_binaria


caminho = 'teste.jpeg'

imagem_cinza = colorida_para_cinza(caminho)
display(imagem_cinza)
#imagem_cinza.save('result_cinza.jpeg')

imagem_binarizada = cinza_para_binario(imagem_cinza)
display(imagem_binarizada)
#imagem_binarizada.save('result_bina.jpeg')
```
